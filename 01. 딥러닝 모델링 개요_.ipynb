# 1.환경준비

* 라이브러리 Import

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import *
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

# 2.Regression : Advertising

## (1) 데이터 전처리

### 1) 데이터 준비

```
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/advertising.csv'
adv = pd.read_csv(path)
adv.head()
```

```
target = 'Sales'
x = adv.drop(target, axis=1)
y = adv.loc[:, target]
```

### 2) 가변수화

```
# 셀의 값들이 범주형이 아니라 숫자형이라 할 필요 없음 !!
```

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
# random_state : 난수 생성기의 시드 역할/ 데이터를 섞거나 샘플링할 때 항상 같은 순서로 섞이게 보장.
```

## (2) ML 연습 : 선형회귀

```
# 선형회귀 알고리즘을 불러 옵시다.
from sklearn.linear_model import LinearRegression
```

### 1) 모델 선언

```
model = LinearRegression()
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
pred = model.predict(x_val)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {root_mean_squared_error(y_val, pred)}')
print(f'MAE   : {mean_absolute_error(y_val, pred)}')
print(f'MAPE  : {mean_absolute_percentage_error(y_val, pred)}')
```

## (3) 딥러닝 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

### 1) 전처리 : Scaling

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)   # .fit: 데이터를 보고 스케일링에 필요한 통계값(최소값, 최대값 등) 을 계산해서 저장.
x_val = scaler.transform(x_val)           # 훈련데이터만 적용 !!
```

### 2) 필요한 함수들 불러오기

```
from keras.models import Sequential
from keras.layers import Dense, Input
from keras.backend import clear_session
```

### 3) 모델 선언

```
x_train.shape
```

```
nfeatures = x_train.shape[1] #num of columns
nfeatures
```

##모델 구조

```
x_train.shape
```

```
# 메모리 정리
clear_session()

# Sequential 타입 모델 선언 (모델 구조!!)
model = Sequential( [Input(shape =(nfeatures,)), # input으로 nfeatures(3)개를 받겠다
                     Dense(1)]  )                # layer 수 = 1

# 모델요약
model.summary()
```

## 컴파일 !!

```
model.compile(optimizer='adam', loss='mse')
```

### 4) 학습

```
model.fit(x_train, y_train)
```

### 5) 예측

```
pred = model.predict(x_val)
```

### 6) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {root_mean_squared_error(y_val, pred)}')
print(f'MAE   : {mean_absolute_error(y_val, pred)}')
print(f'MAPE  : {mean_absolute_percentage_error(y_val, pred)}')
```

# 3.Regression : Carseat

## (1) 데이터 전처리
* 데이터 준비
* 가변수화
* 스케일링(필요하다면)
* 데이터 분할

### 1) 데이터 준비

* 카시트 판매량 데이터

|	변수명	|	설명	|	구분	|
|----|----|----|
|	**Sales** 	|	 **각 지역 판매액(단위 : 1000달러)**	|	**Target**	|
|	CompPrice 	|	지역별 경쟁사 판매가격(달러)	|	feature	|
|	Income 	|	가구당 평균 소득액(1000달러)	|	feature	|
|	Advertising 	|	 각 지역, 회사의 광고 예산(1000달러)	|	feature	|
|	Population 	|	 지역 인구수(단위 : 1000명)	|	feature	|
|	Price 	|	 자사 지역별 판매가격(달러)	|	feature	|
|	ShelveLoc 	|	 진열상태(범주 : Bad, Medium, Good)	|	feature	|
|	Age 	|	 지역 인구의 평균 연령	|	feature	|
|	Education 	|	 교육수준(범주 : 10~18)	|	feature	|
|	Urban 	|	 매장이 도심에 있는지 여부(범주 : Yes, No)	|	feature	|
|	US 	|	 매장이 미국에 있는지 여부(범주 : Yes, No)	|	feature	|


* 데이터 경로 : https://raw.githubusercontent.com/DA4BAM/dataset/master/Carseats.csv

```
path = 'https://raw.githubusercontent.com/DA4BAM/dataset/master/Carseats.csv'
carseat = pd.read_csv(path)
carseat.head()
```

```
target = 'Sales'
x = carseat.drop(target, axis=1)
y = carseat.loc[:, target]
```

### 2) 가변수화

```
cat_cols = ['ShelveLoc', 'Education', 'US', 'Urban']
x = pd.get_dummies(x, columns = cat_cols, drop_first = True)
```

### 3) 데이터분할

```
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=.2, random_state = 20)
```

### 4) Scaling

```
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_val = scaler.transform(x_val)
```

## (2) 모델링
* 필요한 함수들 불러오기
* 모델 선언
* 학습
* 예측
* 성능 검증

```
from keras.models import Sequential
from keras.layers import Dense
from keras.backend import clear_session
```

### 1) 모델 선언

```
x_train.shape
```

```
nfeatures = x_train.shape[1] #num of columns
nfeatures
```

```
# 메모리 정리(필수는 아님!)
clear_session()

# Sequential 타입 모델 선언
model = Sequential( [Input(shape = (nfeatures, )),
                     Dense(1)])

# 모델요약
model.summary()
```

```
model.compile(optimizer='adam', loss='mse')
```

### 2) 학습

```
model.fit(x_train, y_train)
```

### 3) 예측

```
pred = model.predict(x_val)
```

### 4) 검증
만든 모델은 얼마나 정확한지 검증해 봅시다.



```
print(f'RMSE  : {root_mean_squared_error(y_val, pred)}')
print(f'MAE   : {mean_absolute_error(y_val, pred)}')
# print(f'MAPE  : {mean_absolute_percentage_error(y_val, pred)}')
```
